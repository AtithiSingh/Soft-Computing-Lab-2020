#### LINE PLOT

```python
from matplotlib import pyplot as plt 
  
x = [6, 1, 12, 3, 5]  # x-axis values 
   
y = [8, 3, 7, 4, 2] # Y-axis values
  
plt.plot(x,y,color='magenta')  # Function to plot 
   
plt.show() 
```

#### BAR PLOT

```python
from matplotlib import pyplot as plt 
  
x = [6, 1, 12, 3, 5]  # x-axis values 
   
y = [8, 3, 7, 4, 2] # Y-axis values
  
plt.bar(x,y,color='yellow')        # Function to plot 
   
plt.show() 
```

#### HISTOGRAM

```python
from matplotlib import pyplot as plt 
    
y = [8, 3, 7, 4, 2] # Y-axis values
  
plt.hist(y,color='cyan')        # Function to plot 
   
plt.show() 
```

#### SCATTER PLOT

```python
from matplotlib import pyplot as plt 
  
x = [6, 1, 12, 3, 5]  # x-axis values 
   
y = [8, 3, 7, 4, 2] # Y-axis values
  
plt.scatter(x,y,color='red')        # Function to plot 
   
plt.show() 
```
