# Aim: Demonstrate the use of Matplotlib library

### Used to plot various graphs to visualize data

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### 1. Line Plot

```python
# X-axis values 
x=[1,2,3]
# Y-axis values  
y=[5,10,2]
plt.plot(x,y)
plt.show()
```

### Scatter Plot

```python
# X-axis values 
x=[1,2,3,5,8,6]
# Y-axis values  
y=[5,10,2,4,2,3]
plt.scatter(x,y,color='red')
plt.show()
```

### 3. Bar Plot

```python
# X-axis values 
x=[1,2,3,4,5,8,6]
# Y-axis values  
y=[5,10,2,4,2,3,1]
plt.bar(x,y,color='green')
plt.show()
```

### 4. Pie Plot

```python
fruits = ['Guava', 'Mango', 'Strawberry', 'Orange'] 
number = [50, 20, 35, 22] 
# Creating plot 
fig = plt.figure(figsize =(10, 7)) 
plt.pie(number, labels = fruits,autopct='%1.2f%%') 
plt.show() 
```

### 5. Historgram

```python
# X-axis values 
x=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
# Y-axis values  
plt.hist(x,bins=[0,25,50,75,100],color='cyan')
plt.show()
```
