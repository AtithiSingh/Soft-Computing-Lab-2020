 ## Plot the various functions using matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### 1) y = 1 for x<10, y = 0, otherwise



```python
y1 = np.zeros(11)
x1 = np.arange(10,21,1)
y2 = np.ones(11)
x2 = np.arange(0,11,1)
plt.axis(ymin = -1, ymax = 2, xmin = 0, xmax = 20)

plt.plot(x1,y1,'r',label = 'y=0, x>=10')
plt.plot(x2,y2,'g',label = 'y=1, x<10')
plt.plot([10],[0], 'ro')
plt.plot([10],[1],'o',markeredgecolor= 'r',markerfacecolor='none')

plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = "upper left")


plt.show()
```

### 2) y = e^ax for different values of a

- a = 1
- a = 2
- a = -1

```python
# a = 1

x = np.linspace(-2,2,100)
y = np.exp(x)

# setting the axes at the centre
fig = plt.figure(figsize = (8,5))
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')  # for shifting y axis from left to center
ax.spines['bottom'].set_position('zero')  # to position x axis to center
ax.spines['right'].set_color('none')      # to remove the color of right side axes
ax.spines['top'].set_color('none')        # to remove the color of top axis
#ax.xaxis.set_ticks_position('bottom')
#ax.yaxis.set_ticks_position('left')

# plot the function
plt.plot(x,y, 'y', label='y=e^x')
plt.legend(loc='upper left')
plt.xlabel('$x$')
plt.ylabel('$\exp(x)$')

# show the plot
plt.show()
```

```python
# a = 2

x = np.linspace(-2, 2, 100)
y = np.exp(2*x)

fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')


plt.plot(x, y,'y',label = 'y=e^2x')

plt.xlabel('$x$')
plt.ylabel('$\exp(2x)$')

plt.legend(loc = 'upper left')
```

```python
# a = -1

x = np.linspace(-2, 2, 100)
y = np.exp(-1*x)

fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')


plt.plot(x, y,'y',label = 'y=e^-x')

plt.xlabel('$x$')
plt.ylabel('$\exp(-x)$')

plt.legend(loc = 'upper right')
```

### 3) y = 7x^2 + 3x + 10 for 2<= x <= 5

```python
x = np.linspace(2,5,5)
y = []
for i in x:
    y.append(7*i*i + 3*i + 10)

fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1, 1, 1)
#ax.spines['left'].set_position('center')
#ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')


plt.plot(x,y,'r',label = 'y = 7*x*x + 3*x + 10')

plt.legend(loc = 'upper left')

plt.show()
```

### 4) y = 1/(1+e^-x) 


```python
x = np.linspace(-2, 2, 100)
temp = np.exp(-1*x)
y = []
for i in temp:
    y.append(1/(1+i))

fig = plt.figure(figsize = (8,6))

ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')


plt.plot(x, y,'g',label = 'y = 1/(1 + e^-x)')

plt.legend(loc = 'upper left')
```

### 5) y = (1-e^-ax)/(1+e^-ax) for different values of a

- a = 1
- a = 2
- a = -1

```python
# a = 1

x = np.linspace(-2, 2, 100)
temp = np.exp(-1*x)
y = []
for i in temp:
    y.append((1-i)/(1+i))

fig = plt.figure(figsize = (8,6))

ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')


plt.plot(x, y,'b',label = 'y = (1 - e^-x)/(1 + e^-x)')

plt.legend(loc = 'upper left')
```

```python
# a = 2

x = np.linspace(-2, 2, 100)
temp = np.exp(-2*x)
y = []
for i in temp:
    y.append((1-i)/(1+i))

fig = plt.figure(figsize = (8,6))

ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')


plt.plot(x, y,'b',label = 'y = (1 - e^-2x)/(1 + e^-2x)')

plt.legend(loc = 'upper left')
```

```python
# a = -1

x = np.linspace(-2, 2, 100)
temp = np.exp(1*x)
y = []
for i in temp:
    y.append((1-i)/(1+i))

fig = plt.figure(figsize = (8,6))

ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')


plt.plot(x, y,'b',label = 'y = (1 - e^x)/(1 + e^x)')

plt.legend(loc = 'upper right')
```

### 6) y = tan hx



```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.tanh(x)

fig = plt.figure(figsize = (8,6))

ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')


plt.plot(x, y,'g',label = 'y = tan hx')

plt.legend(loc = 'upper left')
```
