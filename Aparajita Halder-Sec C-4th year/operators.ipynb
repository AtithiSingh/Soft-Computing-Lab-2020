```python
#ASSIGNMENT OPERATORS
print("\n ASSIGNMENT OPERATORS")
a=10
c /= a   
print(c)
c //= a 
print(c)
c = 2
c += a   
print(c)
c *= a   
print(c)
c %= a  
print(c)
c *= a  
print(c)

#BITWISE OPERATOR
a = 10
b = 47
print("\n BITWISE OPERATORS")
print(a & b) 
print(a >> 2) 
print(a << 2) 
print(a | b)   
print(~a) 
print(a ^ b) 

#LOGICAL OPERATORS
a = True
b = False 
print("\n LOGICAL OPERATORS")
print(a and b) 
print(a or b) 
print(not a) 
print(not b) 


#ARITHMETIC OPERATORS
x=11
y=12
print("\n ARITHMETIC OPERATORS")
print(x+y)
print(x-y)
print(x*y)
print(x/y)    
print(x//y)   
print(x%y)  
print(x**y) 


#RELATIONAL OPERATORS
x=90
y=100
print("\n RELATIONAL OPERATORS")
print(x>y)
print(x<y)
print(x==y)
print(x!=y)
print(x>=y)
print(x<=y)
     
    
#MEMBERSHIP OPERATORS
l=[1,2,3]
print("\n MEMBERSHIP OPERATORS")
if 3 in l:
    print("IN")
if 5 not in l:
    print("NOT IN")
    
#IDENTITY OPERATORS
x=10
print("\n IDENTITY OPERATORS")
if type(x) is int:
    print("IS")
if type(x) is not float:
    print("IS NOT")
    
```
